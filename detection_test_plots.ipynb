```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from test_detection import test_detection
```

```python
params = np.arange(1.2, 4.05, 0.05)
```

```python
false_positives = []
false_negatives = []
```

```python
for param in tqdm(params):
    _, _, fp, fn = test_detection(10000, param, param) # prominence = max_val = param
    false_positives.append(fp)
    false_negatives.append(fn)
```

```python
plt.plot(params, false_positives, label="false positive")
plt.plot(params, false_negatives, label = "false negative")
plt.xlabel("prominence / max_val")
plt.legend()
```
