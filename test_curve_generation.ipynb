```python
import matplotlib.pyplot as plt
```

```python
from curve_generation import LightCurveGenerator, CurvesGenerator
```

# Example of generating single curve with precise parameters

```python
dt = 500. # In sec
noise_std_coef = 0.3

generator = LightCurveGenerator(noise_std_coef, dt)
```

```python
params = {"number QPEs": 3, 
          "QPE 1": {"start": 1000, "width": 4000, "height": 0.07 - 0.01}, 
          "QPE 2": {"start": 35000, "width": 4000, "height": 0.055 - 0.01},
          "QPE 3": {"start": 68000, "width": 4000, "height": 0.065 - 0.01}
         }
```

```python
signal_duration = 76000 # In sec
noise_level = 0.01

signal, times = generator.generate(noise_level, signal_duration, params)
```

```python
plt.plot(times, signal)
plt.title("My generation")
plt.xlabel("$\Delta$ T (sec)")
plt.ylabel("Net Count Rate (counts/sec)")
```

Original light curve of QPE 1 generated from Chandra data with the same time bin value of 500s can be seen below. My generated curve is quite similar to the real curve 

# Example of generating multiple curves with randomised parameters 

```python
noise_std_coef = 0.3 # Noise std = noise_level * noise_std_coef 
# Right now actual parameters for curves generation can be tweaked in CurvesGenerator._generate_params()
curves_gen = CurvesGenerator(noise_std_coef)
```

```python
res = curves_gen.generate_chunk(10)
```

```python
for i in range(10):
    plt.plot(res[i][1], res[i][0])
    plt.title("My generation")
    plt.xlabel("$\Delta$ T (sec)")
    plt.ylabel("Net Count Rate (counts/sec)")
    print("QPE presence:", res[i][2])
    plt.show()
```
